```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dflist = [] 
for i in range(0,101):
    df = pd.read_csv('result/phi'+str(i*1000)+'.dat', sep=' ', header=None)
    dflist.append([df[0][1:],df[1][1:]])
```

```python
dfarr = np.array(dflist)
```

```python
for i in range(len(dfarr)):
    for j in range(len(dfarr[i][0])):
        dfarr[i][0][j] = float(dfarr[i][0][j])
        dfarr[i][1][j] = float(dfarr[i][1][j])
```
